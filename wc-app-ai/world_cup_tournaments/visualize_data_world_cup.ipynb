```python
plt.figure(figsize=(12,6))

plt.plot(df["Year"], df["Attendance"])

plt.title("Tournament Attendance Over Time")
plt.xlabel("Year")
plt.ylabel("Attendance")

plt.show()
```

```python
df["AttendancePerMatch"] = df["Attendance"] / df["Matches"]

df[["Year", "AttendancePerMatch"]]
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Year"], df["AttendancePerMatch"], marker='o')

plt.title("Attendance Per Match")
plt.xlabel("Year")
plt.ylabel("Fans per Match")

plt.show()
```
