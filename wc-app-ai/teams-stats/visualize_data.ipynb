```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/matches_1930_2022.csv')
```

```python
columns_needed = [
    'home_team',
    'away_team',
    'home_score',
    'away_score',
    'Year',
    'Round',
    'Attendance',
    'Venue',
    'Date'
]

matches = df[columns_needed].copy()
```

```python
matches.drop_duplicates(inplace=True)
```

```python
matches['Date'] = pd.to_datetime(matches['Date'])
```

```python
matches.isnull().sum()
```

```python
def get_result(row):
    if row['home_score'] > row['away_score']:
        return 'Home Win'
    elif row['home_score'] < row['away_score']:
        return 'Away Win'
    else:
        return 'Draw'

matches['result'] = matches.apply(get_result, axis=1)
```

```python
home_df = pd.DataFrame({
    'team': matches['home_team'],
    'opponent': matches['away_team'],
    'goals_for': matches['home_score'],
    'goals_against': matches['away_score'],
    'year': matches['Year'],
    'round': matches['Round'],
    'attendance': matches['Attendance']
})
```

```python
home_df['win'] = (home_df['goals_for'] > home_df['goals_against']).astype(int)

home_df['draw'] = (home_df['goals_for'] == home_df['goals_against']).astype(int)

home_df['loss'] = (home_df['goals_for'] < home_df['goals_against']).astype(int)
```

```python
away_df = pd.DataFrame({
    'team': matches['away_team'],
    'opponent': matches['home_team'],
    'goals_for': matches['away_score'],
    'goals_against': matches['home_score'],
    'year': matches['Year'],
    'round': matches['Round'],
    'attendance': matches['Attendance']
})
```

```python
away_df['win'] = (away_df['goals_for'] > away_df['goals_against']).astype(int)

away_df['draw'] = (away_df['goals_for'] == away_df['goals_against']).astype(int)

away_df['loss'] = (away_df['goals_for'] < away_df['goals_against']).astype(int)
```

```python
teams = pd.concat([home_df, away_df], ignore_index=True)
```

```python
teams.head()
```

```python
team_stats = teams.groupby('team').agg(
    matches_played=('team', 'count'),
    wins=('win', 'sum'),
    draws=('draw', 'sum'),
    losses=('loss', 'sum'),
    goals_scored=('goals_for', 'sum'),
    goals_conceded=('goals_against', 'sum'),
    avg_attendance=('attendance', 'mean')
)
```

```python
team_stats['goal_difference'] = (
    team_stats['goals_scored'] - team_stats['goals_conceded']
)
```

```python
team_stats['win_percentage'] = (
    team_stats['wins'] / team_stats['matches_played'] * 100
).round(2)
```

```python
teams['goal_margin'] = (
    teams['goals_for'] - teams['goals_against']
)
```

```python
biggest_win = teams.loc[
    teams.groupby('team')['goal_margin'].idxmax(),
    ['team', 'opponent', 'goals_for', 'goals_against', 'goal_margin']
]
```

```python
biggest_loss = teams.loc[
    teams.groupby('team')['goal_margin'].idxmin(),
    ['team', 'opponent', 'goals_for', 'goals_against', 'goal_margin']
]
```

```python
most_goals = teams.groupby('team')['goals_for'].max()
```

```python
teams['clean_sheet'] = (
    teams['goals_against'] == 0
).astype(int)

clean_sheets = teams.groupby('team')['clean_sheet'].sum()
```

```python
appearances = teams.groupby('team')['year'].nunique()
```

```python
team_stats['clean_sheets'] = clean_sheets

team_stats['appearances'] = appearances

team_stats['most_goals_in_match'] = most_goals
```
